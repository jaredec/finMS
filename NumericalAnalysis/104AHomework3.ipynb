```python
import math
from numpy import linspace

# Q2a
# First, define Lagrange
# INPUT: x-value, nodes, and degree n,j
def lagrange(x,node,n,j):
    product= 1
    for k in range(n+1):          # equidistributed
        if k!= j:
            product*= (x-node[k])/(node[j]-node[k])
    return product

# Define the Lebesgue function 
def lebesgue(x,node,n):
    sum= 0
    for j in range(n+1):
        sum+= abs(lagrange(x,node,n,j))
    return sum

# Q2b
# Define nodes
def node(n):
    node = []
    for j in range(n+1):
        node.append(-1+j*(2/n))
    return node

ne=1000 # define ne to be 1000

# x-values in [-1,1]
x= linspace(-1, 1, ne)

# y-values in [-1,1]
y1= [lebesgue(i,node(4),4) for i in x]
y2= [lebesgue(i,node(10),10) for i in x]
y3= [lebesgue(i,node(20),20) for i in x]

import matplotlib.pyplot as plt # use to plot

# n=4
plt.plot(x, y1) 
plt.title("L4(x)")
plt.show()

# n=10
plt.plot(x, y2) 
plt.title("L10(x)")
plt.show()

# n=20
plt.plot(x, y3) 
plt.title("L20(x)")
plt.show()

print('When n=4,',' we estimate the Lebesgue constant to be =', max(lebesgue(x,node(4),4)))
print('When n=10,',' we estimate the Lebesgue constant to be =', max(lebesgue(x,node(10),10)))
print('When n=20,',' we estimate the Lebesgue constant to be =', max(lebesgue(x,node(20),20)))

# OUTPUT: Lebesgue function n(x) associated to a given set of pairwise distinct nodes
```

```python
from numpy import prod

# Q3a
# First compute barycentric weights 
def bweights(node):
    l = len(node)
    w = []
    for j in range(l):
        baryw = []
        for k in range(l):
            if k != j:
                baryw.append(node[k]-node[j])
        w.append(1/prod(baryw))
    return w

# Barycentric Polynomial Interpolation, implement weights
# INPUT: x-values, nodes, f evaluated at such nodes
def barycentric(x,node,f):
    l= len(node)
    top= [bweights(node)[i]*f[i]/(x-node[i]) for i in range(l)]
    bottom= [bweights(node)[i]/(x-node[i]) for i in range(l)]
    barysum= sum(top)/sum(bottom)
    return barysum

# Q3b
# Now we approximate f(2)
x= [0, 0.25, 0.52, 0.74, 1.28, 1.50]            # Given values
f= [0, 0.7070, 1.0000, 0.7071, -0.7074, -1]


# Find P5(2)
print("P5(2) = Approximation for f(2)=",barycentric(2,x,f),)

# OUTPUT: P5(2) as an approximation for f(2)
```
